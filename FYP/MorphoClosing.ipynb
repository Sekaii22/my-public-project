```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

Used to ensure that the label mask is gap free

```python
masks_path = "D:/Users/xmyst/Desktop/FYP Materials/Images/before_closing_masks/"
masks = []

# get only files not folders
for file in [f for f in os.listdir(masks_path) if os.path.isfile(os.path.join(masks_path, f))]:
    masks.append(os.path.join(masks_path, file))

masks
```

```python
len(masks)
```

```python
kernel = np.ones((3,3),np.uint8)

for i in range(len(masks)):
    img = cv2.imread(masks[i], cv2.IMREAD_GRAYSCALE)
    closed_img = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
    cv2.imwrite(masks[i], closed_img)
```

## Example Visualization

```python
img = cv2.imread(masks[0], cv2.IMREAD_GRAYSCALE)
kernel = np.ones((3,3), np.uint8)
closed_img = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

fig, ax = plt.subplots(2, 2, figsize=(20,20))
ax[0,0].imshow(img, cmap="gray")
ax[0,0].set_title("Before Closing", fontsize="20")
ax[0,1].imshow(closed_img, cmap="gray")
ax[0,1].set_title("After Closing", fontsize="20")
ax[1,0].hist(img.ravel(), 256, [0,255]);
ax[1,1].hist(closed_img.ravel(), 256, [0,255]);
```
