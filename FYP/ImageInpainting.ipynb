```python
import cv2
import numpy
import os
import matplotlib.pyplot as plt
```

```python
images_path = "D:/Users/xmyst/Desktop/FYP Materials/Images/512x512_images/" # path to image folder
masks_path = "D:/Users/xmyst/Desktop/FYP Materials/Images/512x512_images/inpainting_masks/"
img_inpainted_path = "D:/Users/xmyst/Desktop/FYP Materials/Images/preprocessed_images/" # path to store inpainted images
img_names = ["V1755CG_T", "V1756CG_T", "V1757CG_T", "V1759CG_T", "V1764CG_T", "V1767CG_T"] # names of image to inpaint
images = [] # to store paths of images from folder
masks = []
```

```python
for im in img_names:
    images.append(os.path.join(images_path, im + ".png"))
    masks.append(os.path.join(masks_path, im + "_mask.png"))
```

```python
for i in range(len(images)):
    img = cv2.imread(images[i])
    mask = cv2.imread(masks[i], cv2.IMREAD_GRAYSCALE)
    dst = cv2.inpaint(img, mask, 3, cv2.INPAINT_TELEA)
    cv2.imwrite(img_inpainted_path + img_names[i] + ".png", dst)
```

## Example Visualization

```python
img = cv2.imread('D:/Users/xmyst/Desktop/FYP Materials/Images/512x512_images/V1755CG_T.png')
mask = cv2.imread('D:/Users/xmyst/Desktop/FYP Materials/Images/512x512_images/inpainting_masks/V1755CG_T_mask.png',0)
dst = cv2.inpaint(img,mask,3,cv2.INPAINT_TELEA)
```

```python
img_brg2rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
dst_brg2rgb = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots(1, 2, figsize=(20,10))
ax[0].imshow(img_brg2rgb)
ax[0].set_title("Before", fontsize="25")
ax[1].imshow(dst_brg2rgb)
ax[1].set_title("After", fontsize="25")
```
